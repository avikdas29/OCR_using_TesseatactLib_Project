```
import pytesseract
import shutil
import os
import random
import cv2
from google.colab.patches import cv2_imshow
```

## Open Image:

```
originalImage = cv2.imread('complex_image.png')
cv2_imshow(originalImage)
```

```
extractedInformation = pytesseract.image_to_string(originalImage)
```

```
print(extractedInformation)
```

## Get High Intensity Data

```
grayImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayImage)
```

```
InvertImage = cv2.bitwise_not(grayImage)
cv2_imshow(InvertImage)
```

```
extractedInformation = pytesseract.image_to_string(InvertImage)
extractedInformation
```

## Lets Try Thresholding and get Black and White Image

```
(thresh, blackAndWhiteImage) = cv2.threshold(InvertImage, 80, 255, cv2.THRESH_BINARY)
cv2_imshow(blackAndWhiteImage)
```

```
extractedInformation = pytesseract.image_to_string(blackAndWhiteImage)
print(extractedInformation)
```

## Extracted Colored Data:

```
originalImage = cv2.imread('complex_image.png')
cv2_imshow(originalImage)
```

```
#lets extract red channel from image
red_channel = originalImage[:,:,2]
cv2_imshow(red_channel)
```

```
InvertImage = cv2.bitwise_not(red_channel)
cv2_imshow(InvertImage)
```

```
(thresh, blackAndWhiteImage) = cv2.threshold(InvertImage, 112, 255, cv2.THRESH_BINARY)
cv2_imshow(blackAndWhiteImage)
```

```
extractedInformation = pytesseract.image_to_string(blackAndWhiteImage)
print(extractedInformation)
```

The number in the Red Color Extracted Successfully
